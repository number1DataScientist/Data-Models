```python
X=data.iloc[:,1:]
X
```

```python
x=X.drop('Selling_Price',axis=1)
x
```

```python
y=data.iloc[:,2]
y
```

```python
#from sklearn.datasets import make_regression
#x, y = make_regression(n_samples=10, n_features=1, noise=0.1, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf', C=1.0, gamma='scale')  # You can adjust hyperparameters as needed
svr.fit(X_train_scaled, y_train)
```

```python
y_test
```

```python
#from sklearn.svm import SVR
#SVR=SVR(kernel='linear')
#SVR.fit(x_train,y_train)
```

```python
y_predict=svr.predict(X_test_scaled)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_squared_error=mean_squared_error(y_test,y_predict)
mean_squared_error
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

svr = SVR()

param_grid = {
    'kernel': ['linear', 'rbf'],  
    'C': [0.1, 1, 10],             
    'epsilon': [0.1, 0.2, 0.5]     
}

grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2)
grid_search.fit(X_train_scaled, y_train) 

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from scipy.stats import uniform, randint

svr = SVR()

param_dist = {
    'kernel': ['linear', 'rbf'],                    
    'C': uniform(0.1, 10),                            
    'epsilon': uniform(0.1, 0.5)                      
}

random_search = RandomizedSearchCV(estimator=svr, param_distributions=param_dist, n_iter=10, cv=5, scoring='neg_mean_squared_error', verbose=2, random_state=42)
random_search.fit(X_train_scaled, y_train)  

best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)

best_model = random_search.best_estimator_

y_pred = best_model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
